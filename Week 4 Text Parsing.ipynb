```python
import requests
import bs4
```

Загрузим данные со страницы https://en.wikipedia.org/wiki/Bias-variance_tradeoff 

```python
req = requests.get("https://en.wikipedia.org/wiki/Bias-variance_tradeoff")
```

```python
print req.text
```

Найдем заголовок верхнего уровня, т.е. h1.

```python
parser = bs4.BeautifulSoup(req.text, 'lxml')
```

```python
print parser.findAll('h1')
```

Выводим текст данного заголовка.

```python
x = parser.findAll('h1')
for res in x:
    print res.text
```

Теперь найдем названия всех статей в категории Machine Learning на странице https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms 

Загружаем данные.

```python
req = requests.get("https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms")
```

```python
req.text
```

```python
parser = bs4.BeautifulSoup(req.text, 'lxml')
```

Парсим страницу и находим названия всех статей, выводим также номер статьи в списке. Получили 55 статей, сколько и указано на странице.

```python
p = parser.findAll('div', attrs={'class': "mw-category-group"})
counter = 0
for res in p:
    p1 = res.findAll('a')
    for res1 in p1:
        counter += 1
        print counter, res1.attrs['title']
```
